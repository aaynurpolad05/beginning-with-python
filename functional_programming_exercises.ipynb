```python
#exercise 4
celsius_temp = [0, 10, 20 ,30, 40]
celsius_to_fahranheit= map(lambda c: (c* 9/5)+32, celsius_temp)
print(list(celsius_to_fahranheit))
#with list comprehension
farenheit= [ (c*9/5)+32 for c in celsius_temp]
print(farenheit)
```

```python
#exercise 5
numbers= [1, 15, 22, 37, 44, 59, 60, 72]
new_numbers = filter(lambda num: num%2==0 and num%3==0, numbers)
print(list(new_numbers))
#with list comprehension
devided_numbers= [num for num in numbers if num%2==0 and num%3==0]
print(devided_numbers)
```

```python
#exercise 6
from functools import reduce
num = [1, 2, 3, 4, 5]
summary = reduce(lambda x,y: x+y, num)
print(summary)
```

```python
#exercise 8
words= ["Programming", "is", "fun", "and", "powerful"]
short_words = [ word for word in words if len(word)>5 ]
print(short_words)
```

```python
#exercise 9
from functools import reduce
factorials = {i: reduce(lambda a, b: a * b, range(1, i+1)) for i in range(1, 6)}
print(factorials)

```

```python
def check_password_strenght(password):
    if (password)<8:
        raise ValueError('Password is too short!')
    else:
        print('Password is strong')

try:
    check_password_strenght(7)
except ValueError as e:
    print(f"Error: {e}")
```

```python
def calculate_discount(price, discount_percent):
    assert 0<=discount_percent<=100, "invalid discount percentage"

    final_price = price *( 1- discount_percent /100)
    return final_price
calculate_discount(59,20)
calculate_discount(100,101)
```

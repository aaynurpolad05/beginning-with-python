```python
def calculate_average(*args):
    total = 0
    for i in args:
        total += i
    return total/ len(args)
    
print(calculate_average(1,2,3,4,5))
    
```

```python
def print_profile(**kwargs):
    for k,v in kwargs.items():
        print(f'{k}: {v}')
print_profile(name= 'Alex', age= 28, city = 'New York')
```

```python
def procces_order(customer_name, *items, **details):
    print('Customer:',customer_name)
    print("Items", items)
    print("Details", details)
procces_order('Bob', 'Pizza', 'Coke', delivery_time = '6 PM', adress = '123 Main St')
```

```python
def countdown(start):
    while start>0:
        yield start
        start -=1
for num in countdown(5):
    print(num)
         
```

```python

def lenght(words):
    for word in words:
        yield len(word)
words = ['hello', 'world', 'python', 'is', 'fun']
their_lenghts= lenght(words)
print(next(their_lenghts))
print(next(their_lenghts))
```

```python
def shout(func):
    def wrapper():
        result = func() #string qebul etdiyi hisse
        return result.upper()
    return wrapper
@shout 
def greet():
    return 'Hello, I am a function'
print(greet())
```

```python
#writing and reading files
lines_to_write = ['First line', 'Second line', 'Third line']
with open('my_file.txt','w', encoding = 'utf-8') as f:
    for line in lines_to_write:
        f.write(line + '\n')
with open('my_file.txt', 'r', encoding = 'utf=8') as f:
    content = f.read()
print(content)
```
